# Algoritmo de la Implementación de la transformación en PDFs

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter

def procesar_senal(G=3, a=2, media=1, desviacion=2, N_base=10000, titulo=""):
    """
    Procesa una señal usando dos filtros (LTI y promedio simple) y grafica los histogramas.

    Parámetros:
    - G: tamaño de ventana del filtro de promedio
    - a: parámetro del filtro LTI
    - media: media de la señal de entrada (normal)
    - desviacion: desviación estándar de la señal de entrada
    - N_base: tamaño base de la señal
    - titulo: Título para el gráfico
    """
    N = N_base + G * 100
    X = np.random.normal(media, desviacion, N)

    H = [1/a, 1, 1/a]
    Y = lfilter(H, [1], X)

    filtro_prom = np.ones(G) / G
    Y_filt = lfilter(filtro_prom, [1], Y)

    plt.figure(figsize=(10, 6))
    plt.hist(X, bins=50, density=True, alpha=0.5, label='Entrada (X)')
    plt.hist(Y_filt, bins=50, density=True, alpha=0.5, label='Salida (Y filtrada)')
    plt.legend()
    plt.title(f"Comparación de PDFs entrada/salida\n{titulo}")
    plt.xlabel("Valor")
    plt.ylabel("Densidad")
    plt.grid(True)
    plt.show()

```

```python
procesar_senal(
    G=3,
    a=2,
    media=0,            # Señales digitales pueden centrarse en 0
    desviacion=1,       # Ruido térmico o interferencia
    titulo="Telecomunicaciones: Transmisión Digital en Canales Inalámbricos"
)
```
